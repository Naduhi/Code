```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('./ProcessedData/SLG_Addnoise.csv', sep=",", index_col=0)
plt.plot(Data)
plt.show()
```

```python
mu, sigma = 0, 0.3
noise = np.random.normal(mu, sigma, Data.shape)
New_Data = Data + noise
Data = pd.merge(Data, New_Data, on='Raman_shift')
plt.plot(Data)
plt.show()
```

```python
sigma = 0.1
noise = np.random.normal(mu, sigma, Data.shape)
New_Data = Data + noise
dataframe = pd.merge(Data, New_Data, on='Raman_shift')
plt.plot(dataframe)
plt.show()
```

```python
dataframe.to_csv('./ProcessedData/SLG_Addnoise1.csv')
```
