```python
import numpy as np
import pandas as pd
from math import log, exp
def wgn(x, snr):
    Ps = np.sum(abs(x)**2)/len(x)
    Pn = Ps/(10**((snr/10)))
    noise = np.random.randn(len(x)) * np.sqrt(Pn)
    return noise



# def wgn(audio, snr):
#     #在audio y中 添加噪声 噪声强度SNR为int
#     audio_power = audio ** 2
#     audio_average_power = np.mean(audio_power)
#     audio_average_db = 10 * np.log10(audio_average_power)
#     noise_average_db = audio_average_db - snr
#     noise_average_power = 10 ** (noise_average_db / 10)
#     mean_noise = 0
#     noise = np.random.normal(mean_noise, np.sqrt(noise_average_power), len(audio))
#     return audio + noise
```

```python
file_name = "D:\Data-Processing\SpectraGenerator\\20231111\dataset\\2d_stacked-zer-noise.csv"
data = pd.read_csv(file_name,delimiter=';',skiprows=None,header=None,index_col=None)
df, label = np.array(data.iloc[1:,]),np.array(data.iloc[0,])
```

```python
plt.figure(figsize=(4,2))

plt.xlabel('Raman shift $(cm^{-1})$ ')
plt.ylabel('Raman intensity $(a.u.)$ ')
plt.plot(label, dff.transpose(df),linewidth=0.5)
```

```python

Nums,Ramanshift=df.shape
tmp=np.zeros_like(df[1])
for i in range(Nums):
    noise=wgn(df[i],40)
    new = noise + df[i]
    data_nor = (new - new.min(axis=0))/(new.max(axis=0) - new.min(axis=0))
    tmp=np.vstack((tmp, data_nor))
tmp = np.array(tmp[1:,:])
plt.figure(figsize=(4,2))
plt.plot(label, dff.transpose(tmp),linewidth=0.5)
```
